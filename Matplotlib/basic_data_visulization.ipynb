```python
import matplotlib.pyplot as plt
import numpy as np
```

## Pyplot

```python
xpoint = np.array([0,10])
ypoint = np.array([0,300])
plt.plot(xpoint,ypoint)
plt.show()
```

## Matplotlib Plotting

```python
xpoint = np.array([1,8])
ypoint = np.array([3,10])
plt.plot(xpoint,ypoint)
plt.show()
```

## Plotting Without Line

```python
plt.plot(xpoint, ypoint, 'o')
plt.show()
```

## Multiple Points

```python
# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

xpoint = np.array([1,2,6,8])
ypoint = np.array([3,8,1,10])
plt.plot(xpoint,ypoint)
plt.show()
```

## Default X-Points

```python
# If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc.

ypoint = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoint, marker = "o", linestyle = "dotted", color = "r")
plt.show()
```

## Multiple Lines

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1, x2, y2)
plt.show()
```

## Matplotlib Labels and Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)

plt.plot(x,y)
plt.show()
```

## Matplotlib Adding Grid Lines

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)

plt.plot(x,y)
plt.grid()
plt.show()
```
