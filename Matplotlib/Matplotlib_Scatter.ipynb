# Scatter Plots in Matplotlib

## When to Use Scatter Plots
Scatter plots are ideal for:
- **Exploring Relationships**: Visualize how two numerical variables (e.g., `Unit_Price` vs `Total_Sale`) relate.
- **Identifying Patterns**: Detect trends, clusters, or correlations in data.
- **Spotting Outliers**: Highlight unusual data points (e.g., high `Total_Sale` with low `Unit_Price`).
- **Comparing Groups**: Use colors or markers to compare categories (e.g., `Product_Category`).

## The `plt.scatter()` Function Parameters
The `plt.scatter()` function creates scatter plots. Its syntax is:

```python
plt.scatter(x, y, s=None, c=None, marker=None, alpha=None, label=None)

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
````
